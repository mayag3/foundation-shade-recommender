```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("shades.csv")
print(data.shape)
data.head()
data.fillna(data.mean(),inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
#Normalizing the attributes
features = data[['H', 'S', 'V']]
scaler=MinMaxScaler()
normalized_feature=scaler.fit_transform(features)
normalized_df=pd.DataFrame(normalized_feature,columns=features.columns)


```

```python
#calculating similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix= cosine_similarity(normalized_df)
similarity_df=pd.DataFrame(similarity_matrix,index=data.index,columns=data.index)
similarity_df
```

```python
def recommend_shades(input_hsv, top_n=5):
    input_normalized = scaler.transform([input_hsv])
    similarity_scores = cosine_similarity(normalized_df, input_normalized).flatten()
    top_indices = similarity_scores.argsort()[-top_n:][::-1]
    return data.iloc[top_indices][['brand','product']]

# Get recommendations for NC42 skintone
import matplotlib.colors as mcolors

def hex_to_hsv(hex_code):
    rgb = mcolors.hex2color(hex_code)
    hsv = mcolors.rgb_to_hsv(rgb)
    return hsv
    
hsv = hex_to_hsv('#E0AC69')
nc42_normalized = scaler.transform([hsv])
recommended_shades = recommend_shades(hsv)
print(recommended_shades)
```
